```python
#intro
print("Welcome to Connect Four Game")
A = input("Press >>> 1 <<< To Start Game With Traditional 6 x 7 Board \nPress >>> 2 <<< To Customize The Board \n")
if A == '1':
    column = 6
    row=7
                
elif A == '2' :
    column = int(input("Enter Number of Columnâ€™s for the Custom Board:"))
    row = int(input("Enter Number of Rows for the Custom Board:"))
                
else:
    print('Invalid Input')



class Game_Board():
    
    def __init__(self):
        self.board = [[' ' for x in range(column)] for x in range(row)]
        self.turns=0
        self.last_move=[-1,-1]
        
    def Board_Pattren(self):
        #numbering the column
        print('\n')
        for c in range(column):
            print(f" ({c+1})", end="")
        print("\n")
        #printing the slots pattren of the game board
        for i in range(row):
            print('|',end="")
            for j in range(column):
                print(f" {self.board[i][j]} |", end="")
            print("\n")
        print(f"{'_' * 35}\n")
        
    def choice(self):
        players = ['X','Y']
        return players[self.turns%2]

    def turn(self, column):
        for r in range(row-1,-1,-1):
            if self.board[r][column] == ' ':
                self.board[r][column]= self.choice()
                self.last_move=[r,column]
                self.turns = self.turns + 1
                return True
        return False

    def Bounds(self, r, c):
        return (r >= 0 and r < row and c >= 0 and c < column)

    def check_winner(self):
        last_row = self.last_move[0]
        last_col = self.last_move[1]
        last_letter = self.board[last_row][last_col]

        #[r,c] direction, matching

        directions = [
            [[-1, 0], 0, True],
            [[1, 0], 0, True],
            [[0, -1], 0, True],
            [[0, 1], 0, True],
            [[-1, -1], 0, True],
            [[1, 1], 0, True],
            [[-1, 1], 0, True],
            [[1, -1], 0, True]
        ]
        #Search for matching letters
        for a in range(4):
            for d in directions:
                r = last_row + (d[0][0] * (a+1))
                c = last_col +(d[0][1] * (a+1))

                if d[2] and self.Bounds(r,c) and self.board[r][c] == last_letter:
                    d[1] +=1
                else:
                    #stop searching in this angle
                    d[2] = False

        #Check all direction pairs for 2 pieces in a row
        for i in range(0, 7, 2):
            if(directions[i][1] + directions[i+1][1] >= 3):
                self.Board_Pattren()
                print( "Congratulation! " f"{last_letter} is the winner!")
                return last_letter
        #Did not find a winner
        return False


def play():
    
    game = Game_Board()
    game_over = False
    while not game_over: 
        game.Board_Pattren()

        valid_move = False
        while not valid_move:
            user_move = input(f"{game.choice()}'s turn : ")
            try:
                valid_move = game.turn(int(user_move) -1)
            except:
                print(f"Please enter a number between 1 and {column}")

        game_over = game.check_winner()

        if not any(' ' in x for x in game.board):
            print("The Game Is A Tie")
            return




if __name__=='__main__':
    play()
```
